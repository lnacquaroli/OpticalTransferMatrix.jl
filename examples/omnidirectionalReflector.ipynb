```julia
# Load modules
include("TMMOptics.jl")
include("RIdb.jl")
include("nplot.jl")
using Main.TMMOptics
using Main.RIdb: air, dummy
using PyPlot
```

```julia
# substrate index of refraction: https://refractiveindex.info/?shelf=main&book=NaCl&page=Li
function nacl(x)
    x = x/1000
    n=@. sqrt(1+0.00055+0.19800./(1-(0.050./x).^2)+0.48398./(1-(0.100./x).^2)+0.38696./(1-(0.128./x).^2)+0.25998./(1-(0.158./x).^2)+0.08796./(1-(40.50./x).^2)+3.17064./(1-(60.98./x).^2)+0.30038./(1-(120.34./x).^2))
end
```

```julia
# Define beam
λi = 1000 # intial wavelength [nm]
λf = 15000 # final wavelength [nm]
λ = LinRange(λi, λf, λf-λi+1) # wavelength range [nm]
λ0 = 12000. # reference wavelength
θi = 0
θf = 80
θ = LinRange(θi, θf, θf-θi+1) # angle of incidence [degrees]
p = 1. # polatization (s-wave = 0. and p-wave = 1., or any intermediate)
beam = PlaneWave(λ, λ0, θ, p);
```

```julia
# Define type of layers with their indexes of refraction and thicknesses
l0 = Geometrical(air(beam.λ), 0.)
l1 = Geometrical(dummy(beam.λ, 4.6, 0.), 800.)
l2 = Geometrical(dummy(beam.λ, 1.6, 0.), 1650.)
l3 = Geometrical(nacl(beam.λ), 0.);
```

```julia
# Sequence of materials: elements represent the index position below in the structure (surface down to substrate)
nseq = [l0 l1 l2 l1 l2 l1 l2 l1 l2 l1 l3];
```

```julia
# Perform computation
results = thinfilmoptics(beam, nseq);
```

```julia
# plot the R, T and A spectra
figure()
cf1 = contourf(λ/1e3, θ, Matrix(results.Spectra.Rp'), 40)
cb1_tags = LinRange(minimum(results.Spectra.Rp), maximum(results.Spectra.Rp), 5)
cb1 = colorbar(cf1, ticks=cb1_tags)
cb1[:set_label]("p-Reflectance")
ax1 = gca()
ax1[:tick_params](which="both", direction="in", pad=10, labelsize=14) # ticks offset
axis("tight")
xlabel("Wavelength [um]")
ylabel("Angle of incidence [degree]")
xlim(6,15)
figure()
cf2 = contourf(λ/1e3, θ, Matrix(results.Spectra.Rs'), 40)
cb2_tags = LinRange(minimum(results.Spectra.Rs), maximum(results.Spectra.Rs), 5)
cb2 = colorbar(cf2, ticks=cb2_tags)
cb2[:set_label]("s-Reflectance")
ax2 = gca()
ax2[:tick_params](which="both", direction="in", pad=10, labelsize=14) # ticks offset
axis("tight")
xlabel("Wavelength [um]")
ylabel("Angle of incidence [degree]")
xlim(6,15)
```

```julia
# plot the refractive index profile
nplot(beam.λ, beam.θ, beam.λ0, results.Misc.d, results.Misc.ℓ, results.Field.emf, results.Misc.nλ0, nseq)
```
